# Connect to SQL Database

````python
import pandas as pd 
from sqlalchemy import create_engine
````

````python
# Load and read csv
csv_file = "./data/battles.csv"
got_battles = pd.read_csv(csv_file)

json_file = "./data/character-deaths.json"
character_deaths_uncleaned = pd.read_json(json_file)

csv_file4 = "./data/scripts.csv"
got_scripts = pd.read_csv(csv_file4)

csv_file5 = "./data/ratings.csv"
got_ratings = pd.read_csv(csv_file5)
````

````python
got_ratings = got_ratings.rename(columns ={
                            "Season": "season",
                            "Episode Number": "episode_number", 
                            "Number in Season": "number_in_season",
                            "Episode Name": "episode_name",
                            "Director": "director",
                            "Writer": "writer",
                            "Original Air Date":"original_air_date",
                            "US viewers (million)":"us_viewers_millions",
                            "Runtime (mins)":"runtime_mins",
                            "IMDB Description":"imdb_description",
                            "IMDB votes":"imdb_votes",
                            "Imdb Rating":"imdb_rating",
                            "Notable Death Count":"notable_death_count"
                            })
````

````python
# Establish connection 

connection_string = "postgres:PW@localhost:5432/Game-of-Thrones"
engine = create_engine(f'postgresql://{connection_string}')
````

````python
# Print table names
engine.table_names()

# Load data
got_battles.to_sql(name="got_battles", con=engine, if_exists='append', index=False)
got_ratings.to_sql(name="got_ratings", con=engine, if_exists='append', index=False)


# Read data using Pandas
pd.read_sql_query('select * from got_battles', con=engine)
pd.read_sql_query('select * from got_ratings', con=engine)
````

### SQL Code

```sql

DROP TABLE got_battles;
DROP TABLE got_ratings;

	attacker_1 VARCHAR,

	attacker_3 VARCHAR,
	attacker_4 VARCHAR,
	defender_1 VARCHAR,
	defender_2 VARCHAR,
	defender_3 VARCHAR,
	defender_4 VARCHAR,
	attacker_outcome VARCHAR,
	battle_type VARCHAR,
	major_death INT, 
	major_capture VARCHAR,
	attacker_size INT,
	defender_size INT,
	attacker_commander VARCHAR,
	defender_commander VARCHAR,
	summer VARCHAR,
	location VARCHAR,
	region VARCHAR,
	note VARCHAR
);

CREATE TABLE got_ratings(
	season INT,
	episode_number INT,
	number_in_season INT, 
	episode_name VARCHAR,
	director VARCHAR,
	writer VARCHAR,
	original_air_date VARCHAR,
	us_viewers_millions VARCHAR,

	imdb_description VARCHAR,
	imdb_votes INT,
	imdb_rating INT,
	notable_death_count INT
);

SELECT * FROM got_battles
SELECT * FROM got_ratings
```
